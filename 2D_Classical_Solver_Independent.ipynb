```python
# 2D classical LLF/Rusanov finite-volume solver for reactive Euler + progress variable.
# Independent lab-frame ignition with independent ZND/CJ and overdriven-ZND comparison.

from dataclasses import dataclass
import math
import numpy as np
import matplotlib.pyplot as plt



# Parameters


@dataclass
class Params:
    gamma: float = 1.4
    Q: float = 20.0
    k: float = 1.0
    Ea: float = 20.0
    R: float = 1.0

    rho_a: float = 1.0
    u_a: float = 0.0
    v_a: float = 0.0
    p_a: float = 1.0
    lam_a: float = 0.0

    n_lam: int = 4000
    lam_end: float = 1.0 - 1.0e-6

    D_bracket_low: float = 1.3
    D_bracket_high: float = 7.0
    cj_tol: float = 1.0e-10
    cj_max_iter: int = 80


class ZNDFailure(Exception):
    def __init__(self, lam_fail, last_state=None, message="ZND profile terminated before lam_end"):
        super().__init__(message)
        self.lam_fail = lam_fail
        self.last_state = last_state



# 2D conservative conversion and fluxes


def prim_to_cons_2d(rho, u, v, p, lam, gamma, Q=0.0):
    rho = np.asarray(rho, dtype=float)
    u = np.asarray(u, dtype=float)
    v = np.asarray(v, dtype=float)
    p = np.asarray(p, dtype=float)
    lam = np.asarray(lam, dtype=float)

    z = rho * lam
    E = p / (gamma - 1.0) + 0.5 * rho * (u**2 + v**2) - Q * z

    U = np.empty(rho.shape + (5,), dtype=float)
    U[..., 0] = rho
    U[..., 1] = rho * u
    U[..., 2] = rho * v
    U[..., 3] = E
    U[..., 4] = z
    return U


def cons_to_prim_2d(U, gamma, Q=0.0, check=True):
    rho = U[..., 0]
    mx = U[..., 1]
    my = U[..., 2]
    E = U[..., 3]
    z = U[..., 4]

    if check and np.any(rho <= 0.0):
        raise ValueError(f"Non-positive density: min(rho) = {np.min(rho):.16e}")

    u = mx / rho
    v = my / rho
    lam = z / rho
    kinetic = 0.5 * (mx**2 + my**2) / rho
    p = (gamma - 1.0) * (E - kinetic + Q * z)

    if check and np.any(p <= 0.0):
        raise ValueError(f"Non-positive pressure: min(p) = {np.min(p):.16e}")

    return rho, u, v, p, lam


def temperature_from_prim(rho, p, Rgas=1.0):
    return p / (rho * Rgas)


def reaction_rate_from_prim(rho, p, lam, k, Ea, Rgas):
    T = np.maximum(p / (rho * Rgas), 1.0e-14)
    lam_clip = np.clip(lam, 0.0, 1.0)
    return k * (1.0 - lam_clip) * np.exp(-Ea / (Rgas * T))


def flux_x(U, gamma, Q=0.0):
    rho, u, v, p, lam = cons_to_prim_2d(U, gamma, Q)
    mx = U[..., 1]
    my = U[..., 2]
    E = U[..., 3]
    z = U[..., 4]

    F = np.empty_like(U)
    F[..., 0] = mx
    F[..., 1] = mx * u + p
    F[..., 2] = my * u
    F[..., 3] = u * (E + p)
    F[..., 4] = u * z
    return F


def flux_y(U, gamma, Q=0.0):
    rho, u, v, p, lam = cons_to_prim_2d(U, gamma, Q)
    mx = U[..., 1]
    my = U[..., 2]
    E = U[..., 3]
    z = U[..., 4]

    G = np.empty_like(U)
    G[..., 0] = my
    G[..., 1] = mx * v
    G[..., 2] = my * v + p
    G[..., 3] = v * (E + p)
    G[..., 4] = v * z
    return G


def flux_x_moving(U, gamma, Q, Dref):
    return Dref * U - flux_x(U, gamma, Q)



# ZND reference solver


def reaction_rate_znd(lam, T, par: Params):
    return par.k * (1.0 - lam) * np.exp(-par.Ea / (par.R * T))


def shock_jump(par: Params, D):
    c_a = math.sqrt(par.gamma * par.p_a / par.rho_a)
    q_a = D - par.u_a
    if q_a <= c_a:
        raise ValueError("D is too small: upstream state is not supersonic relative to the shock.")

    M_a = q_a / c_a

    rho_n = par.rho_a * ((par.gamma + 1.0) * M_a**2) / ((par.gamma - 1.0) * M_a**2 + 2.0)
    p_n = par.p_a * (1.0 + 2.0 * par.gamma / (par.gamma + 1.0) * (M_a**2 - 1.0))
    q_n = q_a * par.rho_a / rho_n
    u_n = D - q_n
    lam_n = par.lam_a

    m = rho_n * q_n
    P = p_n + rho_n * q_n**2
    B = par.gamma / (par.gamma - 1.0) * (p_n / rho_n) - par.Q * lam_n + 0.5 * q_n**2

    T_n = p_n / (rho_n * par.R)
    c_n = math.sqrt(par.gamma * p_n / rho_n)

    vn = {
        "rho": rho_n,
        "u": u_n,
        "q": q_n,
        "p": p_n,
        "T": T_n,
        "c": c_n,
        "lam": lam_n,
    }
    inv = {"m": m, "P": P, "B": B}
    return vn, inv


def state_from_lambda(lam, inv, par: Params, D, v_prev):
    m = inv["m"]
    P = inv["P"]
    B = inv["B"]

    a = m**2 * (par.gamma + 1.0) / (2.0 * (par.gamma - 1.0))
    b = -par.gamma * P / (par.gamma - 1.0)
    c = B + par.Q * lam

    disc = b**2 - 4.0 * a * c
    if disc < 0.0:
        raise ZNDFailure(lam_fail=lam, last_state=None)

    sqrt_disc = math.sqrt(disc)
    roots = [(-b - sqrt_disc) / (2.0 * a), (-b + sqrt_disc) / (2.0 * a)]

    candidates = []
    for v_spec in roots:
        if v_spec <= 0.0:
            continue
        rho = 1.0 / v_spec
        q = m * v_spec
        p = P - m**2 * v_spec
        if rho > 0.0 and q > 0.0 and p > 0.0:
            candidates.append(v_spec)

    if not candidates:
        raise ZNDFailure(lam_fail=lam, last_state=None)

    v_spec = min(candidates, key=lambda x: abs(x - v_prev))

    rho = 1.0 / v_spec
    q = m * v_spec
    u = D - q
    p = P - m**2 * v_spec
    T = p / (rho * par.R)
    c_sound = math.sqrt(par.gamma * p / rho)
    M_rel = q / c_sound
    omega = reaction_rate_znd(lam, T, par)

    return {
        "rho": rho,
        "u": u,
        "q": q,
        "p": p,
        "T": T,
        "c": c_sound,
        "M_rel": M_rel,
        "omega": omega,
        "v_spec": v_spec,
    }


def build_znd_profile(par: Params, D):
    vn, inv = shock_jump(par, D)
    lam = np.linspace(par.lam_a, par.lam_end, par.n_lam)

    rho = np.empty_like(lam)
    u = np.empty_like(lam)
    q = np.empty_like(lam)
    p = np.empty_like(lam)
    T = np.empty_like(lam)
    c_sound = np.empty_like(lam)
    M_rel = np.empty_like(lam)
    omega = np.empty_like(lam)
    s = np.empty_like(lam)

    v_prev = 1.0 / vn["rho"]
    last_state = None

    for i, lam_i in enumerate(lam):
        try:
            st = state_from_lambda(lam_i, inv, par, D, v_prev)
        except ZNDFailure:
            raise ZNDFailure(lam_fail=lam_i, last_state=last_state)

        v_prev = st["v_spec"]
        last_state = st

        rho[i] = st["rho"]
        u[i] = st["u"]
        q[i] = st["q"]
        p[i] = st["p"]
        T[i] = st["T"]
        c_sound[i] = st["c"]
        M_rel[i] = st["M_rel"]
        omega[i] = st["omega"]

    ds_dlam = q / omega
    s[0] = 0.0
    for i in range(1, len(lam)):
        dlam = lam[i] - lam[i - 1]
        s[i] = s[i - 1] + 0.5 * (ds_dlam[i] + ds_dlam[i - 1]) * dlam

    s_lam_005 = np.interp(0.05, lam, s)
    s_lam_010 = np.interp(0.10, lam, s)
    s_lam_050 = np.interp(0.50, lam, s)
    s_lam_090 = np.interp(0.90, lam, s)
    s_lam_095 = np.interp(0.95, lam, s)

    return {
        "D": D,
        "s": s,
        "lam": lam,
        "rho": rho,
        "u": u,
        "q": q,
        "p": p,
        "T": T,
        "c": c_sound,
        "M_rel": M_rel,
        "omega": omega,
        "vn_state": vn,
        "invariants": inv,
        "reaction_length_10_90": s_lam_090 - s_lam_010,
        "reaction_length_5_95": s_lam_095 - s_lam_005,
        "reaction_length_total": s[-1],
        "s_lam_05": s_lam_050,
        "s_lam_09": s_lam_090,
        "p_max": np.max(p),
        "T_max": np.max(T),
        "q_end_over_c_end": M_rel[-1],
    }


def try_build_profile(par: Params, D):
    try:
        prof = build_znd_profile(par, D)
        return True, prof, None
    except Exception as e:
        return False, None, e


def find_cj_speed(par: Params):
    c_a = math.sqrt(par.gamma * par.p_a / par.rho_a)
    Dmin = c_a * (1.0 + 1.0e-12)

    D_low = max(par.D_bracket_low, Dmin)
    D_high = max(par.D_bracket_high, 1.1 * D_low)

    ok_low, _, _ = try_build_profile(par, D_low)

    while ok_low and D_low > Dmin * 1.000001:
        D_low = 0.5 * (D_low + Dmin)
        ok_low, _, _ = try_build_profile(par, D_low)

    if ok_low:
        raise RuntimeError("Could not find a failing lower bracket for CJ search.")

    ok_high, prof_high, _ = try_build_profile(par, D_high)
    while not ok_high:
        D_high *= 1.2
        ok_high, prof_high, _ = try_build_profile(par, D_high)

    for _ in range(par.cj_max_iter):
        D_mid = 0.5 * (D_low + D_high)
        ok_mid, prof_mid, _ = try_build_profile(par, D_mid)

        if ok_mid:
            D_high = D_mid
            prof_high = prof_mid
        else:
            D_low = D_mid

        if abs(D_high - D_low) <= par.cj_tol * max(1.0, abs(D_high)):
            break

    return {
        "D_cj": D_high,
        "D_fail": D_low,
        "D_succeed": D_high,
        "profile": prof_high,
    }



# Boundary conditions and 2D LLF/Rusanov operator


def apply_bc_2d(U, x_bc="outflow", left_state=None, right_state=None):
    """One ghost layer in x and y. Periodic in y for all tests."""
    # x-boundaries
    if x_bc == "outflow":
        U[0, :, :] = U[1, :, :]
        U[-1, :, :] = U[-2, :, :]
    elif x_bc == "dirichlet":
        if left_state is None or right_state is None:
            raise ValueError("Dirichlet x_bc requires left_state and right_state.")
        U[0, :, :] = left_state
        U[-1, :, :] = right_state
    else:
        raise ValueError(f"Unknown x_bc = {x_bc}")

    # y-periodic boundaries
    U[:, 0, :] = U[:, -2, :]
    U[:, -1, :] = U[:, 1, :]

    # x-boundary corners again, for consistency
    if x_bc == "dirichlet":
        U[0, :, :] = left_state
        U[-1, :, :] = right_state
    return U


def rhs_2d_llf(U, dx, dy, par: Params, Dref=0.0, moving_x=False,
               x_bc="outflow", left_state=None, right_state=None):
    Uc = U.copy()
    apply_bc_2d(Uc, x_bc=x_bc, left_state=left_state, right_state=right_state)


    UL = Uc[:-1, 1:-1, :]
    UR = Uc[1:, 1:-1, :]

    if moving_x:
        FL = flux_x_moving(UL, par.gamma, par.Q, Dref)
        FR = flux_x_moving(UR, par.gamma, par.Q, Dref)
    else:
        FL = flux_x(UL, par.gamma, par.Q)
        FR = flux_x(UR, par.gamma, par.Q)

    rhoL, uL, vL, pL, _ = cons_to_prim_2d(UL, par.gamma, par.Q)
    rhoR, uR, vR, pR, _ = cons_to_prim_2d(UR, par.gamma, par.Q)
    cL = np.sqrt(par.gamma * pL / rhoL)
    cR = np.sqrt(par.gamma * pR / rhoR)

    if moving_x:
        ax = np.maximum(np.abs(uL - Dref) + cL, np.abs(uR - Dref) + cR)[..., None]
    else:
        ax = np.maximum(np.abs(uL) + cL, np.abs(uR) + cR)[..., None]

    Fhat = 0.5 * (FL + FR) - 0.5 * ax * (UR - UL)


    UB = Uc[1:-1, :-1, :]
    UT = Uc[1:-1, 1:, :]
    GB = flux_y(UB, par.gamma, par.Q)
    GT = flux_y(UT, par.gamma, par.Q)

    rhoB, uB, vB, pB, _ = cons_to_prim_2d(UB, par.gamma, par.Q)
    rhoT, uT, vT, pT, _ = cons_to_prim_2d(UT, par.gamma, par.Q)
    cB = np.sqrt(par.gamma * pB / rhoB)
    cT = np.sqrt(par.gamma * pT / rhoT)
    ay = np.maximum(np.abs(vB) + cB, np.abs(vT) + cT)[..., None]

    Ghat = 0.5 * (GB + GT) - 0.5 * ay * (UT - UB)

    rhs = np.zeros_like(Uc)
    rhs[1:-1, 1:-1, :] = (
        -(Fhat[1:, :, :] - Fhat[:-1, :, :]) / dx
        -(Ghat[:, 1:, :] - Ghat[:, :-1, :]) / dy
    )
    return rhs


def compute_dt_2d(U, dx, dy, par: Params, cfl, Dref=0.0, moving_x=False):
    rho, u, v, p, lam = cons_to_prim_2d(U[1:-1, 1:-1, :], par.gamma, par.Q)
    c = np.sqrt(par.gamma * p / rho)

    if moving_x:
        ax = np.max(np.abs(u - Dref) + c)
    else:
        ax = np.max(np.abs(u) + c)

    ay = np.max(np.abs(v) + c)
    denom = ax / dx + ay / dy
    return cfl / denom


def ssprk3_step_2d(U, dt, dx, dy, par: Params, Dref=0.0, moving_x=False,
                   x_bc="outflow", left_state=None, right_state=None):
    U0 = U.copy()

    K1 = rhs_2d_llf(U0, dx, dy, par, Dref=Dref, moving_x=moving_x,
                    x_bc=x_bc, left_state=left_state, right_state=right_state)
    U1 = U0.copy()
    U1[1:-1, 1:-1, :] = U0[1:-1, 1:-1, :] + dt * K1[1:-1, 1:-1, :]
    apply_bc_2d(U1, x_bc=x_bc, left_state=left_state, right_state=right_state)

    K2 = rhs_2d_llf(U1, dx, dy, par, Dref=Dref, moving_x=moving_x,
                    x_bc=x_bc, left_state=left_state, right_state=right_state)
    U2 = U1.copy()
    U2[1:-1, 1:-1, :] = (
        0.75 * U0[1:-1, 1:-1, :]
        + 0.25 * (U1[1:-1, 1:-1, :] + dt * K2[1:-1, 1:-1, :])
    )
    apply_bc_2d(U2, x_bc=x_bc, left_state=left_state, right_state=right_state)

    K3 = rhs_2d_llf(U2, dx, dy, par, Dref=Dref, moving_x=moving_x,
                    x_bc=x_bc, left_state=left_state, right_state=right_state)
    U3 = U2.copy()
    U3[1:-1, 1:-1, :] = (
        (1.0 / 3.0) * U0[1:-1, 1:-1, :]
        + (2.0 / 3.0) * (U2[1:-1, 1:-1, :] + dt * K3[1:-1, 1:-1, :])
    )
    apply_bc_2d(U3, x_bc=x_bc, left_state=left_state, right_state=right_state)
    return U3


def reaction_step_2d(U, dt, par: Params, x_bc="outflow", left_state=None, right_state=None,
                     react_cfl=0.20, max_substeps=20000):
    Uwork = U.copy()
    apply_bc_2d(Uwork, x_bc=x_bc, left_state=left_state, right_state=right_state)

    remaining = dt
    nsub = 0

    while remaining > 1.0e-15:
        rho, u, v, p, lam = cons_to_prim_2d(Uwork[1:-1, 1:-1, :], par.gamma, par.Q)
        omega = reaction_rate_from_prim(rho, p, lam, par.k, par.Ea, par.R)
        omega_max = np.max(omega)

        if omega_max < 1.0e-14:
            break

        dtau = min(remaining, react_cfl / omega_max)


        z_old = Uwork[1:-1, 1:-1, 4]
        z_new = z_old + dtau * rho * omega
        lam_new = np.clip(z_new / rho, 0.0, 1.0)
        Uwork[1:-1, 1:-1, 4] = rho * lam_new
        remaining -= dtau

        nsub += 1
        if nsub > max_substeps:
            raise RuntimeError("Too many reaction substeps. Try smaller dt or weaker reaction.")

    apply_bc_2d(Uwork, x_bc=x_bc, left_state=left_state, right_state=right_state)
    return Uwork, nsub


def strang_step_2d(U, dt, dx, dy, par: Params, Dref=0.0, moving_x=False,
                   x_bc="outflow", left_state=None, right_state=None, react_cfl=0.20):
    U1, n1 = reaction_step_2d(U, 0.5 * dt, par, x_bc=x_bc,
                              left_state=left_state, right_state=right_state,
                              react_cfl=react_cfl)
    U2 = ssprk3_step_2d(U1, dt, dx, dy, par, Dref=Dref, moving_x=moving_x,
                         x_bc=x_bc, left_state=left_state, right_state=right_state)
    U3, n3 = reaction_step_2d(U2, 0.5 * dt, par, x_bc=x_bc,
                              left_state=left_state, right_state=right_state,
                              react_cfl=react_cfl)
    return U3, n1 + n3



# ZND mapping to 2D grid and diagnostics

def znd_primitives_on_s(s_query, prof, par: Params):
    s_query = np.asarray(s_query)
    rho = np.empty_like(s_query, dtype=float)
    u = np.empty_like(s_query, dtype=float)
    v = np.zeros_like(s_query, dtype=float)
    p = np.empty_like(s_query, dtype=float)
    lam = np.empty_like(s_query, dtype=float)

    mask_up = s_query < 0.0
    mask_dn = ~mask_up

    rho[mask_up] = par.rho_a
    u[mask_up] = par.u_a
    v[mask_up] = par.v_a
    p[mask_up] = par.p_a
    lam[mask_up] = par.lam_a

    s_prof = prof["s"]
    rho[mask_dn] = np.interp(s_query[mask_dn], s_prof, prof["rho"])
    u[mask_dn] = np.interp(s_query[mask_dn], s_prof, prof["u"])
    v[mask_dn] = 0.0
    p[mask_dn] = np.interp(s_query[mask_dn], s_prof, prof["p"])
    lam[mask_dn] = np.interp(s_query[mask_dn], s_prof, prof["lam"])

    return rho, u, v, p, lam


def build_znd_2d_initial_condition(x, y, prof, par: Params):
    X, Y = np.meshgrid(x, y, indexing="ij")
    rho, u, v, p, lam = znd_primitives_on_s(X, prof, par)
    return prim_to_cons_2d(rho, u, v, p, lam, par.gamma, par.Q)


def build_znd_boundary_states(xR, prof, par: Params):
    left = prim_to_cons_2d(
        np.array(par.rho_a), np.array(par.u_a), np.array(par.v_a),
        np.array(par.p_a), np.array(par.lam_a), par.gamma, par.Q
    )
    rhoR, uR, vR, pR, lamR = znd_primitives_on_s(np.array(xR), prof, par)
    right = prim_to_cons_2d(rhoR, uR, vR, pR, lamR, par.gamma, par.Q)
    return left, right


def primitives_y_average(U_inner, par: Params):
    rho, u, v, p, lam = cons_to_prim_2d(U_inner, par.gamma, par.Q)
    T = temperature_from_prim(rho, p, par.R)
    omega = reaction_rate_from_prim(rho, p, lam, par.k, par.Ea, par.R)
    return {
        "rho": np.mean(rho, axis=1),
        "u": np.mean(u, axis=1),
        "v": np.mean(v, axis=1),
        "p": np.mean(p, axis=1),
        "T": np.mean(T, axis=1),
        "lam": np.mean(lam, axis=1),
        "omega": np.mean(omega, axis=1),
        "rho_full": rho,
        "u_full": u,
        "v_full": v,
        "p_full": p,
        "T_full": T,
        "lam_full": lam,
    }


def znd_reference_profile_on_x(x, prof, par: Params):
    rho, u, v, p, lam = znd_primitives_on_s(x, prof, par)
    T = temperature_from_prim(rho, p, par.R)
    omega = reaction_rate_from_prim(rho, p, lam, par.k, par.Ea, par.R)
    return {"rho": rho, "u": u, "v": v, "p": p, "T": T, "lam": lam, "omega": omega}


def integrate_l1(x, err):
    return np.trapezoid(np.abs(err), x)


def integrate_l2(x, err):
    return math.sqrt(np.trapezoid(err**2, x))


def error_report_on_x(x, avg, ref, smooth_window=(1.0, 80.0)):
    masks = {
        "full": np.ones_like(x, dtype=bool),
        "smooth": (x >= smooth_window[0]) & (x <= smooth_window[1]),
    }
    out = {}
    for name, mask in masks.items():
        xm = x[mask]
        for var in ["p", "rho", "u", "T", "lam"]:
            err = avg[var][mask] - ref[var][mask]
            out[f"{var}_{name}"] = {
                "L1": integrate_l1(xm, err),
                "L2": integrate_l2(xm, err),
                "Linf": np.max(np.abs(err)),
            }
    return out


def planarity_diagnostics(U_inner, par: Params):
    rho, u, v, p, lam = cons_to_prim_2d(U_inner, par.gamma, par.Q)
    return {
        "max_abs_v": np.max(np.abs(v)),
        "max_std_rho_y": np.max(np.std(rho, axis=1)),
        "max_std_u_y": np.max(np.std(u, axis=1)),
        "max_std_v_y": np.max(np.std(v, axis=1)),
        "max_std_p_y": np.max(np.std(p, axis=1)),
        "max_std_lam_y": np.max(np.std(lam, axis=1)),
        "lambda_min": np.min(lam),
        "lambda_max": np.max(lam),
        "rho_min": np.min(rho),
        "p_min": np.min(p),
    }


def linf_drift_from_initial(U_final, U_initial, par: Params):
    f = primitives_y_average(U_final, par)
    i = primitives_y_average(U_initial, par)
    return {
        "p": np.max(np.abs(f["p"] - i["p"])),
        "rho": np.max(np.abs(f["rho"] - i["rho"])),
        "u": np.max(np.abs(f["u"] - i["u"])),
        "T": np.max(np.abs(f["T"] - i["T"])),
        "lam": np.max(np.abs(f["lam"] - i["lam"])),
        "v": np.max(np.abs(f["v"] - i["v"])),
    }



# homogeneous reaction-source sanity test


def solve_reaction_source_0d_2d(par: Params, final_time=10.0, react_cfl=0.20,
                                Nx=64, Ny=16, sample_times=None,
                                rho0=1.0, u0=0.0, v0=0.0, p0=20.0, lam0=0.0):
    if sample_times is None:
        sample_times = [0.0, 1.0, 2.5, 5.0, 7.5, final_time]
    sample_times = sorted(sample_times)

    U = np.zeros((Nx + 2, Ny + 2, 5))
    rho = rho0 * np.ones((Nx, Ny))
    u = u0 * np.ones((Nx, Ny))
    v = v0 * np.ones((Nx, Ny))
    p = p0 * np.ones((Nx, Ny))
    lam = lam0 * np.ones((Nx, Ny))
    U[1:-1, 1:-1, :] = prim_to_cons_2d(rho, u, v, p, lam, par.gamma, par.Q)
    apply_bc_2d(U, x_bc="outflow")

    t = 0.0
    next_id = 0
    history = []
    total_substeps = 0
    U_initial = U[1:-1, 1:-1, :].copy()

    def record(tt):
        avg = primitives_y_average(U[1:-1, 1:-1, :], par)
        history.append((tt, np.mean(avg["lam"]), np.mean(avg["p"]), np.mean(avg["T"]), np.mean(avg["omega"])))

    while next_id < len(sample_times) and abs(sample_times[next_id]) < 1e-15:
        record(0.0)
        next_id += 1

    while t < final_time - 1e-14:
        target = sample_times[next_id] if next_id < len(sample_times) else final_time
        dt = target - t
        U, nsub = reaction_step_2d(U, dt, par, x_bc="outflow", react_cfl=react_cfl)
        total_substeps += nsub
        t = target
        if next_id < len(sample_times):
            record(t)
            next_id += 1
        else:
            break

    avg = primitives_y_average(U[1:-1, 1:-1, :], par)
    rho_f, u_f, v_f, p_f, lam_f = cons_to_prim_2d(U[1:-1, 1:-1, :], par.gamma, par.Q)
    T_f = temperature_from_prim(rho_f, p_f, par.R)
    E_drift = np.max(np.abs(U[1:-1, 1:-1, 3] - U_initial[..., 3]))
    rho_drift = np.max(np.abs(U[1:-1, 1:-1, 0] - U_initial[..., 0]))

    diagnostics = {
        "rho_mean": np.mean(rho_f),
        "p_mean": np.mean(p_f),
        "T_mean": np.mean(T_f),
        "lam_mean": np.mean(lam_f),
        "omega_mean": np.mean(avg["omega"]),
        "rho_min": np.min(rho_f),
        "p_min": np.min(p_f),
        "T_min": np.min(T_f),
        "lam_min": np.min(lam_f),
        "lam_max": np.max(lam_f),
        "max_abs_u": np.max(np.abs(u_f)),
        "max_abs_v": np.max(np.abs(v_f)),
        "max_std_p": np.max(np.std(p_f, axis=1)),
        "max_std_T": np.max(np.std(T_f, axis=1)),
        "max_std_lam": np.max(np.std(lam_f, axis=1)),
        "E_Linf_drift": E_drift,
        "rho_Linf_drift": rho_drift,
        "reaction_substeps": total_substeps,
    }
    return np.array(history), U[1:-1, 1:-1, :], diagnostics



# 2D planar CJ-ZND preservation solver


def solve_2d_planar_znd_preservation(par: Params, prof, xL=-20.0, xR=120.0,
                                     yL=0.0, yR=20.0, Nx=800, Ny=16,
                                     final_time=10.0, cfl=0.20, react_cfl=0.20,
                                     snapshot_times=None):
    if snapshot_times is None:
        snapshot_times = [0.0, 1.0, 2.5, final_time]
    snapshot_times = sorted(snapshot_times)

    dx = (xR - xL) / Nx
    dy = (yR - yL) / Ny
    x = xL + (np.arange(Nx) + 0.5) * dx
    y = yL + (np.arange(Ny) + 0.5) * dy

    left_state, right_state = build_znd_boundary_states(xR, prof, par)

    U = np.zeros((Nx + 2, Ny + 2, 5))
    U[1:-1, 1:-1, :] = build_znd_2d_initial_condition(x, y, prof, par)
    apply_bc_2d(U, x_bc="dirichlet", left_state=left_state, right_state=right_state)

    U_initial = U[1:-1, 1:-1, :].copy()

    t = 0.0
    steps = 0
    reaction_substeps = 0
    snapshots = []
    next_snap = 0

    while next_snap < len(snapshot_times) and abs(snapshot_times[next_snap]) < 1e-15:
        snapshots.append((0.0, U[1:-1, 1:-1, :].copy()))
        next_snap += 1

    while t < final_time - 1.0e-14:
        dt = compute_dt_2d(U, dx, dy, par, cfl, Dref=prof["D"], moving_x=True)
        if t + dt > final_time:
            dt = final_time - t
        if next_snap < len(snapshot_times):
            target = snapshot_times[next_snap]
            if t < target <= t + dt:
                dt = target - t

        U, nsub = strang_step_2d(
            U, dt, dx, dy, par,
            Dref=prof["D"], moving_x=True,
            x_bc="dirichlet", left_state=left_state, right_state=right_state,
            react_cfl=react_cfl
        )
        reaction_substeps += nsub
        t += dt
        steps += 1

        if next_snap < len(snapshot_times):
            target = snapshot_times[next_snap]
            if abs(t - target) < 1.0e-12 or t > target:
                snapshots.append((t, U[1:-1, 1:-1, :].copy()))
                next_snap += 1

    out = {
        "x": x,
        "y": y,
        "dx": dx,
        "dy": dy,
        "U": U[1:-1, 1:-1, :].copy(),
        "U_initial": U_initial,
        "snapshots": snapshots,
        "steps": steps,
        "reaction_substeps": reaction_substeps,
    }
    return out



# Plotting


def plot_step3_history(history):
    fig, axes = plt.subplots(4, 1, figsize=(9, 10), sharex=True)
    t = history[:, 0]
    axes[0].plot(t, history[:, 1], marker="o")
    axes[1].plot(t, history[:, 2], marker="o")
    axes[2].plot(t, history[:, 3], marker="o")
    axes[3].plot(t, history[:, 4], marker="o")
    axes[0].set_ylabel(r"$\langle\lambda\rangle$")
    axes[1].set_ylabel(r"$\langle p\rangle$")
    axes[2].set_ylabel(r"$\langle T\rangle$")
    axes[3].set_ylabel(r"$\langle\omega\rangle$")
    axes[3].set_xlabel("t")
    fig.suptitle("Step 3: spatially homogeneous 2D reaction-source test")
    plt.tight_layout()
    plt.show()


def plot_znd_preservation_profiles(x, avg, ref, title_suffix="", xlim=(-20.0, 120.0)):
    fig, axes = plt.subplots(5, 1, figsize=(9, 14), sharex=True)
    for ax, var, lab in zip(
        axes,
        ["p", "rho", "u", "T", "lam"],
        ["p", r"$\rho$", "u", "T", r"$\lambda$"],
    ):
        ax.plot(x, ref[var], "k-", lw=2.0, label="ZND")
        ax.plot(x, avg[var], "r--", lw=1.5, label="2D PDE y-average")
        ax.set_ylabel(lab)
        ax.grid(False)
    axes[0].legend()
    axes[-1].set_xlabel("s")
    for ax in axes:
        ax.set_xlim(xlim)
    fig.suptitle(f"Step 4: 2D planar ZND preservation {title_suffix}")
    plt.tight_layout()
    plt.show()


def plot_znd_preservation_near_zone(x, avg, ref, title_suffix=""):
    plot_znd_preservation_profiles(x, avg, ref, title_suffix=title_suffix + ", near reaction zone", xlim=(-5.0, 80.0))


def plot_step4_lambda_snapshots(x, snapshots, prof, par: Params, xlim=(-5.0, 80.0)):
    plt.figure(figsize=(9, 4.8))
    lam_ref = znd_reference_profile_on_x(x, prof, par)["lam"]
    for t, U in snapshots:
        avg = primitives_y_average(U, par)
        plt.plot(x, avg["lam"], lw=1.2, label=f"t={t:.3f}")
    plt.plot(x, lam_ref, "k--", lw=2.0, label="ZND ref")
    plt.xlabel("s")
    plt.ylabel(r"$\lambda$")
    plt.xlim(xlim)
    plt.title("Step 4: ZND-preservation, lambda snapshots")
    plt.legend()
    plt.tight_layout()
    plt.show()


def plot_step4_2d_fields(x, y, U, par: Params, xlim=(-5.0, 80.0)):
    rho, u, v, p, lam = cons_to_prim_2d(U, par.gamma, par.Q)
    T = temperature_from_prim(rho, p, par.R)
    fields = [(p, "p"), (rho, r"$\rho$"), (u, "u"), (v, "v"), (T, "T"), (lam, r"$\lambda$")]

    mask = (x >= xlim[0]) & (x <= xlim[1])
    xplot = x[mask]

    fig, axes = plt.subplots(len(fields), 1, figsize=(9, 14), sharex=True)
    for ax, (field, title) in zip(axes, fields):
        data = field[mask, :].T
        im = ax.imshow(
            data,
            origin="lower",
            aspect="auto",
            extent=[xplot[0], xplot[-1], y[0], y[-1]],
        )
        ax.set_ylabel("y")
        ax.set_title(title)
        fig.colorbar(im, ax=ax)
    axes[-1].set_xlabel("s")
    fig.suptitle("Step 4 final 2D fields: planar ZND should remain y-uniform")
    plt.tight_layout()
    plt.show()




# transversely perturbed planar CJ-ZND solver


def periodic_distance(y, y0, period):

    return (y - y0 + 0.5 * period) % period - 0.5 * period


def build_perturbed_znd_2d_initial_condition(x, y, prof, par: Params, perturbation=None):

    if perturbation is None:
        perturbation = {}

    mode = int(perturbation.get("mode", 1))
    shock_amp = float(perturbation.get("shock_amp", 0.50))
    p_amp = float(perturbation.get("p_amp", 0.01))
    v_amp = float(perturbation.get("v_amp", 0.02))
    center_s = float(perturbation.get("center_s", prof.get("s_lam_05", 15.0)))
    sigma_s = float(perturbation.get("sigma_s", 8.0))
    phase = float(perturbation.get("phase", 0.0))
    phase_v = float(perturbation.get("phase_v", 0.5 * math.pi))

    X, Y = np.meshgrid(x, y, indexing="ij")
    dy = y[1] - y[0] if len(y) > 1 else 1.0
    Ly = y[-1] - y[0] + dy
    y0 = y[0]

    theta = 2.0 * math.pi * mode * (Y - y0) / Ly + phase
    eta = shock_amp * np.sin(theta)


    for item in perturbation.get("extra_modes", []):
        if isinstance(item, dict):
            m = int(item.get("mode", 2))
            A = float(item.get("amp", 0.0))
            ph = float(item.get("phase", 0.0))
        else:
            m = int(item[0])
            A = float(item[1])
            ph = float(item[2]) if len(item) > 2 else 0.0
        eta += A * np.sin(2.0 * math.pi * m * (Y - y0) / Ly + ph)

    s_eff = X - eta
    rho, u, v, p, lam = znd_primitives_on_s(s_eff, prof, par)


    if sigma_s > 0.0:
        window_s = np.exp(-0.5 * ((X - center_s) / sigma_s) ** 2)
    else:
        window_s = np.ones_like(X)

    if p_amp != 0.0:
        p = p * np.maximum(1.0 + p_amp * np.sin(theta) * window_s, 1.0e-8)

    if v_amp != 0.0:
        v = v + v_amp * np.sin(theta + phase_v) * window_s


    for hs in perturbation.get("hotspots", []):
        cs = float(hs.get("center_s", center_s))
        cy = float(hs.get("center_y", y0 + 0.5 * Ly))
        ss = float(hs.get("sigma_s", 5.0))
        sy = float(hs.get("sigma_y", 0.15 * Ly))
        p_hot = float(hs.get("p_amp", 0.0))
        v_hot = float(hs.get("v_amp", 0.0))
        lam_hot = float(hs.get("lam_amp", 0.0))

        dy_per = periodic_distance(Y, cy, Ly)
        gs = np.exp(-0.5 * ((X - cs) / ss) ** 2) if ss > 0.0 else np.ones_like(X)
        gy = np.exp(-0.5 * (dy_per / sy) ** 2) if sy > 0.0 else np.ones_like(Y)
        g = gs * gy

        if p_hot != 0.0:
            p = p * np.maximum(1.0 + p_hot * g, 1.0e-8)
        if v_hot != 0.0:

            v = v + v_hot * (dy_per / max(sy, 1.0e-12)) * g
        if lam_hot != 0.0:
            lam = lam + lam_hot * g

    lam = np.clip(lam, 0.0, 1.0)
    return prim_to_cons_2d(rho, u, v, p, lam, par.gamma, par.Q)

def front_positions_by_level(x, field_xy, level, increasing=True):

    Nx, Ny = field_xy.shape
    xpos = np.full(Ny, np.nan)

    for j in range(Ny):
        f = field_xy[:, j]
        if increasing:
            idx = np.where((f[:-1] <= level) & (f[1:] > level))[0]
        else:
            idx = np.where((f[:-1] >= level) & (f[1:] < level))[0]
        if len(idx) == 0:
            continue
        i = idx[0]
        f0, f1 = f[i], f[i + 1]
        if abs(f1 - f0) < 1.0e-14:
            xpos[j] = x[i]
        else:
            xpos[j] = x[i] + (level - f0) * (x[i + 1] - x[i]) / (f1 - f0)
    return xpos


def finite_front_stats(xpos):
    good = np.isfinite(xpos)
    if np.count_nonzero(good) < 2:
        return {"mean": np.nan, "std": np.nan, "amp": np.nan, "min": np.nan, "max": np.nan, "valid_count": int(np.count_nonzero(good))}
    xf = xpos[good]
    return {
        "mean": float(np.mean(xf)),
        "std": float(np.std(xf)),
        "amp": float(0.5 * (np.max(xf) - np.min(xf))),
        "min": float(np.min(xf)),
        "max": float(np.max(xf)),
        "valid_count": int(np.count_nonzero(good)),
    }


def transverse_diagnostics(x, U_inner, par: Params, prof=None):
    rho, u, v, p, lam = cons_to_prim_2d(U_inner, par.gamma, par.Q)
    T = temperature_from_prim(rho, p, par.R)

    out = {
        "max_abs_v": float(np.max(np.abs(v))),
        "rms_v": float(np.sqrt(np.mean(v**2))),
        "max_std_p_y": float(np.max(np.std(p, axis=1))),
        "max_std_rho_y": float(np.max(np.std(rho, axis=1))),
        "max_std_u_y": float(np.max(np.std(u, axis=1))),
        "max_std_T_y": float(np.max(np.std(T, axis=1))),
        "max_std_lam_y": float(np.max(np.std(lam, axis=1))),
        "lambda_min": float(np.min(lam)),
        "lambda_max": float(np.max(lam)),
        "rho_min": float(np.min(rho)),
        "p_min": float(np.min(p)),
    }

    if prof is not None:
        p_threshold = 0.5 * (par.p_a + prof["vn_state"]["p"])
        x_p = front_positions_by_level(x, p, p_threshold, increasing=True)
        x_lam = front_positions_by_level(x, lam, 0.5, increasing=True)
        ps = finite_front_stats(x_p)
        ls = finite_front_stats(x_lam)
        out.update({
            "shock_front_mean": ps["mean"],
            "shock_front_std": ps["std"],
            "shock_front_amp": ps["amp"],
            "shock_front_valid": ps["valid_count"],
            "lambda05_front_mean": ls["mean"],
            "lambda05_front_std": ls["std"],
            "lambda05_front_amp": ls["amp"],
            "lambda05_front_valid": ls["valid_count"],
        })
    return out




def compute_schlieren_like(x, y, U_inner, par: Params):

    rho, u, v, p, lam = cons_to_prim_2d(U_inner, par.gamma, par.Q)
    drho_ds = np.gradient(rho, x, axis=0, edge_order=2)
    drho_dy = np.gradient(rho, y, axis=1, edge_order=2)
    grad = np.sqrt(drho_ds**2 + drho_dy**2)
    chi = np.log1p(grad)
    return chi, grad


def pressure_temperature_fields(U_inner, par: Params):
    rho, u, v, p, lam = cons_to_prim_2d(U_inner, par.gamma, par.Q)
    T = temperature_from_prim(rho, p, par.R)
    return p, T


def update_extreme_histories(U_inner, par: Params, p_max, T_max):
    p, T = pressure_temperature_fields(U_inner, par)
    np.maximum(p_max, p, out=p_max)
    np.maximum(T_max, T, out=T_max)
    return p_max, T_max

def solve_2d_perturbed_znd(par: Params, prof, xL=-20.0, xR=120.0,
                           yL=0.0, yR=20.0, Nx=500, Ny=32,
                           final_time=8.0, cfl=0.20, react_cfl=0.20,
                           perturbation=None, snapshot_times=None,
                           diagnostic_dt=0.25):
    if snapshot_times is None:
        snapshot_times = [0.0, 2.0, 4.0, final_time]
    snapshot_times = sorted(snapshot_times)

    if diagnostic_dt is not None and diagnostic_dt > 0.0:
        diagnostic_times = list(np.arange(0.0, final_time + 0.5 * diagnostic_dt, diagnostic_dt))
        if abs(diagnostic_times[-1] - final_time) > 1.0e-12:
            diagnostic_times.append(final_time)
    else:
        diagnostic_times = [0.0, final_time]
    diagnostic_times = sorted(set([round(t, 14) for t in diagnostic_times if t <= final_time + 1e-12]))

    dx = (xR - xL) / Nx
    dy = (yR - yL) / Ny
    x = xL + (np.arange(Nx) + 0.5) * dx
    y = yL + (np.arange(Ny) + 0.5) * dy

    left_state, right_state = build_znd_boundary_states(xR, prof, par)

    U = np.zeros((Nx + 2, Ny + 2, 5))
    U[1:-1, 1:-1, :] = build_perturbed_znd_2d_initial_condition(x, y, prof, par, perturbation)
    apply_bc_2d(U, x_bc="dirichlet", left_state=left_state, right_state=right_state)

    U_initial = U[1:-1, 1:-1, :].copy()
    p0, T0 = pressure_temperature_fields(U_initial, par)
    p_max = p0.copy()
    T_max = T0.copy()

    t = 0.0
    steps = 0
    reaction_substeps = 0
    snapshots = []
    history = []
    next_snap = 0
    next_diag = 0

    def record_diag(tt):
        Uinner = U[1:-1, 1:-1, :]
        d = transverse_diagnostics(x, Uinner, par, prof=prof)
        chi, grad_rho = compute_schlieren_like(x, y, Uinner, par)
        d["schlieren_max"] = float(np.max(chi))
        d["grad_rho_max"] = float(np.max(grad_rho))
        d["pmax_global"] = float(np.max(p_max))
        d["Tmax_global"] = float(np.max(T_max))
        d["t"] = float(tt)
        history.append(d)

    while next_snap < len(snapshot_times) and abs(snapshot_times[next_snap]) < 1e-15:
        snapshots.append((0.0, U[1:-1, 1:-1, :].copy()))
        next_snap += 1
    while next_diag < len(diagnostic_times) and abs(diagnostic_times[next_diag]) < 1e-15:
        record_diag(0.0)
        next_diag += 1

    while t < final_time - 1.0e-14:
        dt = compute_dt_2d(U, dx, dy, par, cfl, Dref=prof["D"], moving_x=True)
        if t + dt > final_time:
            dt = final_time - t

        if next_snap < len(snapshot_times):
            target = snapshot_times[next_snap]
            if t < target <= t + dt:
                dt = target - t
        if next_diag < len(diagnostic_times):
            target = diagnostic_times[next_diag]
            if t < target <= t + dt:
                dt = min(dt, target - t)

        U, nsub = strang_step_2d(
            U, dt, dx, dy, par,
            Dref=prof["D"], moving_x=True,
            x_bc="dirichlet", left_state=left_state, right_state=right_state,
            react_cfl=react_cfl
        )
        reaction_substeps += nsub
        update_extreme_histories(U[1:-1, 1:-1, :], par, p_max, T_max)
        t += dt
        steps += 1

        while next_snap < len(snapshot_times) and (abs(t - snapshot_times[next_snap]) < 1e-12 or t > snapshot_times[next_snap]):
            snapshots.append((snapshot_times[next_snap], U[1:-1, 1:-1, :].copy()))
            next_snap += 1
        while next_diag < len(diagnostic_times) and (abs(t - diagnostic_times[next_diag]) < 1e-12 or t > diagnostic_times[next_diag]):
            record_diag(diagnostic_times[next_diag])
            next_diag += 1

    return {
        "x": x,
        "y": y,
        "dx": dx,
        "dy": dy,
        "U": U[1:-1, 1:-1, :].copy(),
        "U_initial": U_initial,
        "snapshots": snapshots,
        "history": history,
        "steps": steps,
        "reaction_substeps": reaction_substeps,
        "p_max": p_max.copy(),
        "T_max": T_max.copy(),
        "schlieren": compute_schlieren_like(x, y, U[1:-1, 1:-1, :], par)[0],
        "perturbation": perturbation if perturbation is not None else {},
    }

def print_step5_history_table(history, stride=4):
    print("\nStep 5/6 transverse-mode and cellular-diagnostic history:")
    print("-" * 132)
    print(
        f"{'t':>8s} {'max_std_p_y':>14s} {'max_std_lam_y':>16s} "
        f"{'max|v|':>12s} {'shock_amp':>12s} {'lam05_amp':>12s} {'schl_max':>12s} {'pmax':>12s}"
    )
    print("-" * 132)
    rows = history[::max(1, stride)]
    if history and rows[-1] is not history[-1]:
        rows = rows + [history[-1]]
    for d in rows:
        print(
            f"{d['t']:8.3f} {d['max_std_p_y']:14.6e} {d['max_std_lam_y']:16.6e} "
            f"{d['max_abs_v']:12.6e} {d.get('shock_front_amp', np.nan):12.6e} "
            f"{d.get('lambda05_front_amp', np.nan):12.6e} {d.get('schlieren_max', np.nan):12.6e} "
            f"{d.get('pmax_global', np.nan):12.6e}"
        )
    print("-" * 132)

def plot_step5_yaverage_profiles(x, avg, ref, title_suffix="", xlim=(-20.0, 120.0)):
    fig, axes = plt.subplots(5, 1, figsize=(9, 14), sharex=True)
    for ax, var, lab in zip(
        axes,
        ["p", "rho", "u", "T", "lam"],
        ["p", r"$\rho$", "u", "T", r"$\lambda$"],
    ):
        ax.plot(x, ref[var], "k-", lw=2.0, label="CJ-ZND")
        ax.plot(x, avg[var], "r--", lw=1.5, label="2D PDE y-average")
        ax.set_ylabel(lab)
        ax.grid(False)
    axes[0].legend()
    axes[-1].set_xlabel("s")
    for ax in axes:
        ax.set_xlim(xlim)
    fig.suptitle(f"Step 5: perturbed ZND, y-averaged profile {title_suffix}")
    plt.tight_layout()
    plt.show()


def plot_step5_transverse_history(history):
    if not history:
        return
    t = np.array([d["t"] for d in history])
    max_std_p = np.array([d["max_std_p_y"] for d in history])
    max_std_lam = np.array([d["max_std_lam_y"] for d in history])
    max_abs_v = np.array([d["max_abs_v"] for d in history])
    shock_amp = np.array([d.get("shock_front_amp", np.nan) for d in history])
    lam_amp = np.array([d.get("lambda05_front_amp", np.nan) for d in history])

    fig, axes = plt.subplots(4, 1, figsize=(9, 10), sharex=True)
    axes[0].plot(t, max_std_p, marker="o", ms=3)
    axes[1].plot(t, max_std_lam, marker="o", ms=3)
    axes[2].plot(t, max_abs_v, marker="o", ms=3)
    axes[3].plot(t, shock_amp, marker="o", ms=3, label="p-threshold shock front")
    axes[3].plot(t, lam_amp, marker="s", ms=3, label=r"$\lambda=0.5$ front")
    axes[0].set_ylabel(r"$\max_s \sigma_y(p)$")
    axes[1].set_ylabel(r"$\max_s \sigma_y(\lambda)$")
    axes[2].set_ylabel(r"$\max |v|$")
    axes[3].set_ylabel("front amplitude")
    axes[3].set_xlabel("t")
    axes[3].legend()
    fig.suptitle("Step 5: transverse perturbation diagnostics")
    plt.tight_layout()
    plt.show()


def plot_step5_front_positions(x, y, snapshots, par: Params, prof, xlim=(-5.0, 80.0)):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4.8), sharey=True)
    p_threshold = 0.5 * (par.p_a + prof["vn_state"]["p"])
    for t, U in snapshots:
        rho, u, v, p, lam = cons_to_prim_2d(U, par.gamma, par.Q)
        xp = front_positions_by_level(x, p, p_threshold, increasing=True)
        xl = front_positions_by_level(x, lam, 0.5, increasing=True)
        axes[0].plot(xp, y, lw=1.2, label=f"t={t:.2f}")
        axes[1].plot(xl, y, lw=1.2, label=f"t={t:.2f}")
    axes[0].set_xlabel("s-front")
    axes[1].set_xlabel("s-front")
    axes[0].set_ylabel("y")
    axes[0].set_title("p-threshold shock front")
    axes[1].set_title(r"reaction front $\lambda=0.5$")
    for ax in axes:
        ax.set_xlim(xlim)
        ax.legend(fontsize=8)
    fig.suptitle("Step 5: transverse front positions")
    plt.tight_layout()
    plt.show()


def plot_step5_2d_fields(x, y, U, par: Params, xlim=(-5.0, 80.0)):
    rho, u, v, p, lam = cons_to_prim_2d(U, par.gamma, par.Q)
    T = temperature_from_prim(rho, p, par.R)
    fields = [(p, "p"), (rho, r"$\rho$"), (u, "u"), (v, "v"), (T, "T"), (lam, r"$\lambda$")]

    mask = (x >= xlim[0]) & (x <= xlim[1])
    xplot = x[mask]

    fig, axes = plt.subplots(len(fields), 1, figsize=(9, 14), sharex=True)
    for ax, (field, title) in zip(axes, fields):
        data = field[mask, :].T
        im = ax.imshow(
            data,
            origin="lower",
            aspect="auto",
            extent=[xplot[0], xplot[-1], y[0], y[-1]],
        )
        ax.set_ylabel("y")
        ax.set_title(title)
        fig.colorbar(im, ax=ax)
    axes[-1].set_xlabel("s")
    fig.suptitle("Step 5/6 final 2D fields: transversely perturbed ZND")
    plt.tight_layout()
    plt.show()



def plot_step6_cellular_diagnostics(x, y, U, p_max, T_max, schlieren, par: Params, xlim=(-5.0, 90.0)):

    rho, u, v, p, lam = cons_to_prim_2d(U, par.gamma, par.Q)
    T = temperature_from_prim(rho, p, par.R)
    fields = [
        (schlieren, r"Schlieren-like $\log(1+|\nabla\rho|)$"),
        (p_max, r"soot-foil analogue $p_{\max}(s,y)$"),
        (T_max, r"$T_{\max}(s,y)$"),
        (p, "final p"),
        (T, "final T"),
        (lam, r"final $\lambda$"),
        (v, "final v"),
    ]

    mask = (x >= xlim[0]) & (x <= xlim[1])
    xplot = x[mask]

    fig, axes = plt.subplots(len(fields), 1, figsize=(10, 16), sharex=True)
    for ax, (field, title) in zip(axes, fields):
        data = field[mask, :].T
        im = ax.imshow(
            data,
            origin="lower",
            aspect="auto",
            extent=[xplot[0], xplot[-1], y[0], y[-1]],
        )
        ax.set_ylabel("y")
        ax.set_title(title)
        fig.colorbar(im, ax=ax)
    axes[-1].set_xlabel("s")
    fig.suptitle("Step 6 diagnostics: shocks, pressure maxima, and transverse structure")
    plt.tight_layout()
    plt.show()


def plot_step6_schlieren_and_sootfoil(x, y, schlieren, p_max, xlim=(-5.0, 90.0)):
    mask = (x >= xlim[0]) & (x <= xlim[1])
    xplot = x[mask]
    fig, axes = plt.subplots(2, 1, figsize=(10, 7.5), sharex=True)
    for ax, field, title in zip(
        axes,
        [schlieren, p_max],
        [r"Schlieren-like $\log(1+|\nabla\rho|)$", r"soot-foil analogue $p_{\max}(s,y)$"],
    ):
        im = ax.imshow(field[mask, :].T, origin="lower", aspect="auto", extent=[xplot[0], xplot[-1], y[0], y[-1]])
        ax.set_ylabel("y")
        ax.set_title(title)
        fig.colorbar(im, ax=ax)
    axes[-1].set_xlabel("s")
    fig.suptitle("Step 6 cellular-pattern diagnostics")
    plt.tight_layout()
    plt.show()


def print_dict_table(title, dct):
    print(title)
    print("-" * 72)
    for k, v in dct.items():
        print(f"{k:>28s} = {v:.8e}" if isinstance(v, (float, np.floating)) else f"{k:>28s} = {v}")
    print("-" * 72)


def print_error_block(title, err):
    print(title)
    print("-" * 88)
    print(f"{'var':>8s} {'region':>10s} {'L1':>18s} {'L2':>18s} {'Linf':>18s}")
    print("-" * 88)
    for region in ["full", "smooth"]:
        for var in ["p", "rho", "u", "T", "lam"]:
            e = err[f"{var}_{region}"]
            print(f"{var:>8s} {region:>10s} {e['L1']:18.8e} {e['L2']:18.8e} {e['Linf']:18.8e}")
    print("-" * 88)




# Main


def _fourier_shift_y(Y, y0, Ly, amp=0.0, mode=1, phase=0.0, extra_modes=None):
    eta = amp * np.sin(2.0 * math.pi * int(mode) * (Y - y0) / Ly + phase)
    if extra_modes is None:
        extra_modes = []
    for item in extra_modes:
        if isinstance(item, dict):
            m = int(item.get("mode", 2))
            A = float(item.get("amp", 0.0))
            ph = float(item.get("phase", 0.0))
        else:
            m = int(item[0])
            A = float(item[1])
            ph = float(item[2]) if len(item) > 2 else 0.0
        eta += A * np.sin(2.0 * math.pi * m * (Y - y0) / Ly + ph)
    return eta


def initialize_independent_ignition_2d(x, y, par: Params, ignition=None):

    if ignition is None:
        ignition = {}

    X, Y = np.meshgrid(x, y, indexing="ij")
    dx = x[1] - x[0] if len(x) > 1 else 1.0
    dy = y[1] - y[0] if len(y) > 1 else 1.0
    Ly = y[-1] - y[0] + dy
    y0 = y[0]

    rho = par.rho_a * np.ones_like(X)
    u = par.u_a * np.ones_like(X)
    v = par.v_a * np.ones_like(X)
    p = par.p_a * np.ones_like(X)
    lam = par.lam_a * np.ones_like(X)

    x_left = float(ignition.get("x_ign_left", x[0]))
    x_right0 = float(ignition.get("x_ign_right", 0.0))
    rho_ign = float(ignition.get("rho_ign", par.rho_a))
    u_ign = float(ignition.get("u_ign", 0.0))
    v_ign = float(ignition.get("v_ign", 0.0))
    p_ign = float(ignition.get("p_ign", 55.0))
    lam_ign = float(ignition.get("lam_ign", 0.0))

.
    edge_amp = float(ignition.get("edge_amp", 0.0))
    edge_mode = int(ignition.get("edge_mode", 1))
    edge_phase = float(ignition.get("edge_phase", 0.0))
    edge_extra_modes = ignition.get("edge_extra_modes", [])
    eta = _fourier_shift_y(Y, y0, Ly, amp=edge_amp, mode=edge_mode,
                           phase=edge_phase, extra_modes=edge_extra_modes)
    x_right = x_right0 + eta

    smooth = bool(ignition.get("smooth", False))
    smooth_width = float(ignition.get("smooth_width", max(dx, 1.0e-12)))

    if smooth:

        left_w = 0.5 * (1.0 + np.tanh((X - x_left) / smooth_width))
        right_w = 0.5 * (1.0 - np.tanh((X - x_right) / smooth_width))
        w = left_w * right_w
        rho = rho + w * (rho_ign - rho)
        u = u + w * (u_ign - u)
        v = v + w * (v_ign - v)
        p = p + w * (p_ign - p)
        lam = lam + w * (lam_ign - lam)
    else:
        mask = (X >= x_left) & (X <= x_right)
        rho[mask] = rho_ign
        u[mask] = u_ign
        v[mask] = v_ign
        p[mask] = p_ign
        lam[mask] = lam_ign


    p_amp = float(ignition.get("p_amp", 0.0))
    v_amp = float(ignition.get("v_amp", 0.0))
    seed_sigma_x = float(ignition.get("seed_sigma_x", 6.0))
    seed_center_x = float(ignition.get("seed_center_x", x_right0))
    seed_mode = int(ignition.get("seed_mode", edge_mode))
    seed_phase = float(ignition.get("seed_phase", 0.0))
    seed_phase_v = float(ignition.get("seed_phase_v", 0.5 * math.pi))
    theta = 2.0 * math.pi * seed_mode * (Y - y0) / Ly + seed_phase
    if seed_sigma_x > 0.0:
        wx = np.exp(-0.5 * ((X - seed_center_x) / seed_sigma_x)**2)
    else:
        wx = np.ones_like(X)

    if p_amp != 0.0:
        p = p * np.maximum(1.0 + p_amp * np.sin(theta) * wx, 1.0e-8)
    if v_amp != 0.0:
        v = v + v_amp * np.sin(theta + seed_phase_v) * wx


    for hs in ignition.get("hotspots", []):
        cx = float(hs.get("center_x", x_right0))
        cy = float(hs.get("center_y", y0 + 0.5 * Ly))
        sx = float(hs.get("sigma_x", 5.0))
        sy = float(hs.get("sigma_y", 0.15 * Ly))
        p_hot = float(hs.get("p_amp", 0.0))
        v_hot = float(hs.get("v_amp", 0.0))
        lam_hot = float(hs.get("lam_amp", 0.0))

        dy_per = periodic_distance(Y, cy, Ly)
        gx = np.exp(-0.5 * ((X - cx) / sx)**2) if sx > 0.0 else np.ones_like(X)
        gy = np.exp(-0.5 * (dy_per / sy)**2) if sy > 0.0 else np.ones_like(Y)
        g = gx * gy

        if p_hot != 0.0:
            p = p * np.maximum(1.0 + p_hot * g, 1.0e-8)
        if v_hot != 0.0:
            v = v + v_hot * (dy_per / max(sy, 1.0e-12)) * g
        if lam_hot != 0.0:
            lam = lam + lam_hot * g

    lam = np.clip(lam, 0.0, 1.0)
    p = np.maximum(p, 1.0e-12)
    rho = np.maximum(rho, 1.0e-12)
    return prim_to_cons_2d(rho, u, v, p, lam, par.gamma, par.Q)


def front_positions_by_level_lab(x, field_xy, level, direction="decreasing", choose="rightmost"):

    Nx, Ny = field_xy.shape
    xpos = np.full(Ny, np.nan)

    for j in range(Ny):
        f = field_xy[:, j]
        if direction == "decreasing":
            idx = np.where((f[:-1] >= level) & (f[1:] < level))[0]
        elif direction == "increasing":
            idx = np.where((f[:-1] <= level) & (f[1:] > level))[0]
        else:
            idx = np.where(((f[:-1] - level) * (f[1:] - level)) <= 0.0)[0]

        if len(idx) == 0:
            continue
        i = idx[-1] if choose == "rightmost" else idx[0]
        f0, f1 = f[i], f[i + 1]
        if abs(f1 - f0) < 1.0e-14:
            xpos[j] = x[i]
        else:
            xpos[j] = x[i] + (level - f0) * (x[i + 1] - x[i]) / (f1 - f0)
    return xpos


def lab_front_diagnostics(x, U_inner, par: Params, p_level=2.0, lambda_level=0.5):
    rho, u, v, p, lam = cons_to_prim_2d(U_inner, par.gamma, par.Q)
    T = temperature_from_prim(rho, p, par.R)

    x_p = front_positions_by_level_lab(x, p, p_level, direction="decreasing", choose="rightmost")
    x_lam = front_positions_by_level_lab(x, lam, lambda_level, direction="decreasing", choose="rightmost")
    ps = finite_front_stats(x_p)
    ls = finite_front_stats(x_lam)

    return {
        "max_abs_v": float(np.max(np.abs(v))),
        "rms_v": float(np.sqrt(np.mean(v**2))),
        "max_std_p_y": float(np.max(np.std(p, axis=1))),
        "max_std_rho_y": float(np.max(np.std(rho, axis=1))),
        "max_std_u_y": float(np.max(np.std(u, axis=1))),
        "max_std_T_y": float(np.max(np.std(T, axis=1))),
        "max_std_lam_y": float(np.max(np.std(lam, axis=1))),
        "lambda_min": float(np.min(lam)),
        "lambda_max": float(np.max(lam)),
        "rho_min": float(np.min(rho)),
        "p_min": float(np.min(p)),
        "shock_front_mean": ps["mean"],
        "shock_front_std": ps["std"],
        "shock_front_amp": ps["amp"],
        "shock_front_valid": ps["valid_count"],
        "lambda05_front_mean": ls["mean"],
        "lambda05_front_std": ls["std"],
        "lambda05_front_amp": ls["amp"],
        "lambda05_front_valid": ls["valid_count"],
    }


def solve_2d_independent_ignition(par: Params,
                                  xL=-50.0, xR=250.0,
                                  yL=0.0, yR=30.0,
                                  Nx=800, Ny=80,
                                  final_time=25.0,
                                  cfl=0.20, react_cfl=0.20,
                                  ignition=None,
                                  snapshot_times=None,
                                  tracking_dt=0.25,
                                  p_track_level=2.0,
                                  lambda_track_level=0.5):
    if snapshot_times is None:
        snapshot_times = [0.0, 5.0, 10.0, final_time]
    snapshot_times = sorted(snapshot_times)

    dx = (xR - xL) / Nx
    dy = (yR - yL) / Ny
    x = xL + (np.arange(Nx) + 0.5) * dx
    y = yL + (np.arange(Ny) + 0.5) * dy

    U = np.zeros((Nx + 2, Ny + 2, 5))
    U[1:-1, 1:-1, :] = initialize_independent_ignition_2d(x, y, par, ignition)
    apply_bc_2d(U, x_bc="outflow")

    U_initial = U[1:-1, 1:-1, :].copy()
    p0, T0 = pressure_temperature_fields(U_initial, par)
    p_max = p0.copy()
    T_max = T0.copy()

    t = 0.0
    steps = 0
    reaction_substeps = 0
    snapshots = []
    trajectory = []
    history = []
    next_snap = 0
    next_track = 0.0

    def record_track(tt):
        Uinner = U[1:-1, 1:-1, :]
        rho, u, v, p, lam = cons_to_prim_2d(Uinner, par.gamma, par.Q)
        x_p = front_positions_by_level_lab(x, p, p_track_level, direction="decreasing", choose="rightmost")
        x_lam = front_positions_by_level_lab(x, lam, lambda_track_level, direction="decreasing", choose="rightmost")
        ps = finite_front_stats(x_p)
        ls = finite_front_stats(x_lam)
        d = lab_front_diagnostics(x, Uinner, par, p_level=p_track_level, lambda_level=lambda_track_level)
        chi, grad = compute_schlieren_like(x, y, Uinner, par)
        d["t"] = float(tt)
        d["schlieren_max"] = float(np.max(chi))
        d["grad_rho_max"] = float(np.max(grad))
        d["pmax_global"] = float(np.max(p_max))
        d["Tmax_global"] = float(np.max(T_max))
        history.append(d)
        trajectory.append({
            "t": float(tt),
            "x_shock_mean": ps["mean"],
            "x_shock_std": ps["std"],
            "x_shock_amp": ps["amp"],
            "x_shock_valid": ps["valid_count"],
            "x_lambda_mean": ls["mean"],
            "x_lambda_std": ls["std"],
            "x_lambda_amp": ls["amp"],
            "x_lambda_valid": ls["valid_count"],
        })

    while next_snap < len(snapshot_times) and abs(snapshot_times[next_snap]) < 1e-15:
        snapshots.append((0.0, U[1:-1, 1:-1, :].copy()))
        next_snap += 1
    record_track(0.0)
    next_track += tracking_dt

    while t < final_time - 1.0e-14:
        dt = compute_dt_2d(U, dx, dy, par, cfl, Dref=0.0, moving_x=False)
        if t + dt > final_time:
            dt = final_time - t

        if next_snap < len(snapshot_times):
            target = snapshot_times[next_snap]
            if t < target <= t + dt:
                dt = target - t
        if t < next_track <= t + dt:
            dt = next_track - t

        U, nsub = strang_step_2d(
            U, dt, dx, dy, par,
            Dref=0.0,
            moving_x=False,
            x_bc="outflow",
            react_cfl=react_cfl,
        )
        reaction_substeps += nsub
        update_extreme_histories(U[1:-1, 1:-1, :], par, p_max, T_max)

        t += dt
        steps += 1

        if t >= next_track - 1.0e-12:
            record_track(t)
            next_track += tracking_dt

        while next_snap < len(snapshot_times) and (abs(t - snapshot_times[next_snap]) < 1e-12 or t > snapshot_times[next_snap]):
            snapshots.append((snapshot_times[next_snap], U[1:-1, 1:-1, :].copy()))
            next_snap += 1

    return {
        "x": x,
        "y": y,
        "dx": dx,
        "dy": dy,
        "U": U[1:-1, 1:-1, :].copy(),
        "U_initial": U_initial,
        "snapshots": snapshots,
        "trajectory": trajectory,
        "history": history,
        "steps": steps,
        "reaction_substeps": reaction_substeps,
        "p_max": p_max.copy(),
        "T_max": T_max.copy(),
        "schlieren": compute_schlieren_like(x, y, U[1:-1, 1:-1, :], par)[0],
        "ignition": ignition if ignition is not None else {},
        "p_track_level": p_track_level,
        "lambda_track_level": lambda_track_level,
    }


def trajectory_to_arrays(trajectory):
    keys = ["t", "x_shock_mean", "x_lambda_mean", "x_shock_amp", "x_lambda_amp",
            "x_shock_std", "x_lambda_std", "x_shock_valid", "x_lambda_valid"]
    out = {k: np.array([d.get(k, np.nan) for d in trajectory], dtype=float) for k in keys}
    return out


def fit_mean_front_speed_2d(trajectory, t_min=10.0, front="shock"):
    a = trajectory_to_arrays(trajectory)
    t = a["t"]
    if front == "shock":
        x = a["x_shock_mean"]
    else:
        x = a["x_lambda_mean"]
    mask = np.isfinite(x) & (t >= t_min)
    if np.count_nonzero(mask) < 2:
        return np.nan, np.nan, np.nan
    coeff = np.polyfit(t[mask], x[mask], 1)
    D = coeff[0]
    b = coeff[1]
    resid = x[mask] - (D * t[mask] + b)
    rms = float(np.sqrt(np.mean(resid**2)))
    return float(D), float(b), rms


def znd_profile_on_s_grid(prof, s_grid, par: Params):
    rho, u, v, p, lam = znd_primitives_on_s(s_grid, prof, par)
    T = temperature_from_prim(rho, p, par.R)
    return {
        "s": np.asarray(s_grid),
        "rho": rho,
        "u": u,
        "v": v,
        "p": p,
        "T": T,
        "lam": lam,
    }


def _interp_nan(x, f, xq):
    xq = np.asarray(xq)
    out = np.full_like(xq, np.nan, dtype=float)
    mask = (xq >= x[0]) & (xq <= x[-1])
    if np.any(mask):
        out[mask] = np.interp(xq[mask], x, f)
    return out


def extract_mean_shock_attached_profile(x, U_inner, par: Params, x_shock_mean,
                                        s_max=120.0, n_points=1800):
    s = np.linspace(0.0, s_max, n_points)
    avg = primitives_y_average(U_inner, par)
    xq = x_shock_mean - s
    out = {"s": s}
    for name in ["rho", "u", "v", "p", "T", "lam"]:
        out[name] = _interp_nan(x, avg[name], xq)
    return out


def extract_local_shock_attached_profile(x, y, U_inner, par: Params, x_shock_y,
                                         s_max=120.0, n_points=1800):
    s = np.linspace(0.0, s_max, n_points)
    rho, u, v, p, lam = cons_to_prim_2d(U_inner, par.gamma, par.Q)
    T = temperature_from_prim(rho, p, par.R)
    fields = {"rho": rho, "u": u, "v": v, "p": p, "T": T, "lam": lam}
    out = {"s": s}
    good = np.isfinite(x_shock_y)

    for name, arr in fields.items():
        vals = []
        for j in range(len(y)):
            if not good[j]:
                continue
            xq = x_shock_y[j] - s
            vals.append(_interp_nan(x, arr[:, j], xq))
        if len(vals) == 0:
            out[name] = np.full_like(s, np.nan)
        else:
            out[name] = np.nanmean(np.vstack(vals), axis=0)
    return out


def reaction_zone_metrics_from_profile(profile):
    s = profile["s"]
    lam = profile["lam"]
    mask = np.isfinite(s) & np.isfinite(lam)
    if np.count_nonzero(mask) < 3:
        return {"s_lam_05": np.nan, "s_lam_09": np.nan, "L10_90": np.nan, "L5_95": np.nan}
    ss = s[mask]
    ll = lam[mask]
    order = np.argsort(ll)
    ll = ll[order]
    ss = ss[order]

    keep = np.concatenate(([True], np.diff(ll) > 1.0e-12))
    ll = ll[keep]
    ss = ss[keep]
    def sval(level):
        if len(ll) < 2 or level < ll[0] or level > ll[-1]:
            return np.nan
        return float(np.interp(level, ll, ss))
    s05 = sval(0.5)
    s09 = sval(0.9)
    s10 = sval(0.1)
    s90 = sval(0.9)
    s05p = sval(0.05)
    s95 = sval(0.95)
    return {
        "s_lam_05": s05,
        "s_lam_09": s09,
        "L10_90": s90 - s10 if np.isfinite(s90) and np.isfinite(s10) else np.nan,
        "L5_95": s95 - s05p if np.isfinite(s95) and np.isfinite(s05p) else np.nan,
    }


def profile_error_report_on_s(num, ref, smooth_window=(1.0, 80.0)):
    s = num["s"]
    out = {}
    full = np.isfinite(s)
    sm = full & (s >= smooth_window[0]) & (s <= smooth_window[1])
    for name in ["p", "rho", "u", "T", "lam"]:
        diff = np.abs(num[name] - ref[name])
        finite = np.isfinite(diff)
        for region, mask0 in [("full", full), ("smooth", sm)]:
            mask = mask0 & finite
            if np.count_nonzero(mask) < 2:
                vals = {"L1": np.nan, "L2": np.nan, "Linf": np.nan}
            else:
                vals = {
                    "L1": integrate_l1(s[mask], diff[mask]),
                    "L2": integrate_l2(s[mask], diff[mask]),
                    "Linf": float(np.max(diff[mask])),
                }
            out[f"{name}_{region}"] = vals
    return out


def print_profile_metrics(title, metrics):
    print(title)
    print("-" * 72)
    for k, v in metrics.items():
        print(f"{k:>24s} = {v:.8e}")
    print("-" * 72)


def print_step7_trajectory_table(trajectory, stride=4):
    print("\nStep 7 front/tranverse diagnostic history:")
    print("-" * 132)
    print(f"{'t':>9s} {'x_shock':>12s} {'x_lam05':>12s} {'A_shock':>12s} {'A_lam05':>12s} {'max_std_p':>14s} {'max_std_lam':>14s} {'max|v|':>12s}")
    print("-" * 132)
    for i, d in enumerate(trajectory):
        if i % stride != 0 and i != len(trajectory) - 1:
            continue

        print(
            f"{d.get('t', np.nan):9.4f} "
            f"{d.get('x_shock_mean', np.nan):12.6f} "
            f"{d.get('x_lambda_mean', np.nan):12.6f} "
            f"{d.get('x_shock_amp', np.nan):12.6f} "
            f"{d.get('x_lambda_amp', np.nan):12.6f} "
            f"{'':>14s} {'':>14s} {'':>12s}"
        )
    print("-" * 132)


def print_step7_history_table(history, stride=4):
    print("\nStep 7 transverse/cellular diagnostic history:")
    print("-" * 132)
    print(f"{'t':>9s} {'x_shock':>12s} {'x_lam05':>12s} {'A_shock':>12s} {'A_lam05':>12s} {'max_std_p':>14s} {'max_std_lam':>14s} {'max|v|':>12s}")
    print("-" * 132)
    for i, d in enumerate(history):
        if i % stride != 0 and i != len(history) - 1:
            continue
        print(
            f"{d.get('t', np.nan):9.4f} "
            f"{d.get('shock_front_mean', np.nan):12.6f} "
            f"{d.get('lambda05_front_mean', np.nan):12.6f} "
            f"{d.get('shock_front_amp', np.nan):12.6f} "
            f"{d.get('lambda05_front_amp', np.nan):12.6f} "
            f"{d.get('max_std_p_y', np.nan):14.6e} "
            f"{d.get('max_std_lam_y', np.nan):14.6e} "
            f"{d.get('max_abs_v', np.nan):12.6e}"
        )
    print("-" * 132)


def plot_step7_lab_snapshots(x, snapshots, par: Params, xlim=None):
    fig, axes = plt.subplots(5, 1, figsize=(10, 14), sharex=True)
    names = ["p", "rho", "u", "T", "lam"]
    labels = ["p", r"$\rho$", "u", "T", r"$\lambda$"]
    for t, Uinner in snapshots:
        avg = primitives_y_average(Uinner, par)
        for ax, name in zip(axes, names):
            ax.plot(x, avg[name], label=f"t={t:.2f}")
    for ax, lab in zip(axes, labels):
        ax.set_ylabel(lab)
        if xlim is not None:
            ax.set_xlim(*xlim)
    axes[0].set_title("Step 7: independent 2D lab-frame ignition, y-averaged evolution")
    axes[-1].set_xlabel("x")
    axes[0].legend(loc="best")
    plt.tight_layout()
    plt.show()


def plot_step7_front_trajectory(trajectory, D_cj, t_fit_min=10.0):
    a = trajectory_to_arrays(trajectory)
    t = a["t"]
    xs = a["x_shock_mean"]
    xl = a["x_lambda_mean"]
    Ds, bs, rms_s = fit_mean_front_speed_2d(trajectory, t_min=t_fit_min, front="shock")
    Dl, bl, rms_l = fit_mean_front_speed_2d(trajectory, t_min=t_fit_min, front="lambda")

    plt.figure(figsize=(10, 5))
    plt.plot(t, xs, "o-", ms=3, label="mean p-threshold shock front")
    plt.plot(t, xl, "s-", ms=3, label=r"mean reaction front $\lambda=0.5$")
    tf = np.linspace(np.nanmin(t), np.nanmax(t), 200)
    if np.isfinite(Ds):
        plt.plot(tf, Ds * tf + bs, "k--", label=f"shock fit D={Ds:.4f}")
    if np.isfinite(Dl):
        plt.plot(tf, Dl * tf + bl, "r--", label=f"lambda fit D={Dl:.4f}")
    plt.plot(tf, D_cj * tf + (np.nanmean(xs) - D_cj * np.nanmean(t)), "g:", label=f"CJ guide D={D_cj:.4f}")
    plt.axvline(t_fit_min, color="0.5", ls=":", lw=1.0)
    plt.xlabel("t")
    plt.ylabel("mean front position")
    plt.title("Step 7: independent PDE front tracking")
    plt.legend(loc="best")
    plt.grid(True, alpha=0.25)
    plt.tight_layout()
    plt.show()


def plot_step7_shock_attached_profiles(pde_mean, pde_local, znd_cj, znd_speed=None,
                                       title_suffix="", xlim=(0.0, 100.0)):
    fig, axes = plt.subplots(5, 1, figsize=(10, 14), sharex=True)
    names = ["p", "rho", "u", "T", "lam"]
    labels = ["p", r"$\rho$", "u", "T", r"$\lambda$"]
    for ax, name, lab in zip(axes, names, labels):
        ax.plot(znd_cj["s"], znd_cj[name], "k-", lw=2.0, label="CJ-ZND")
        if znd_speed is not None:
            ax.plot(znd_speed["s"], znd_speed[name], "b-.", lw=1.5, label="ZND at D_PDE")
        ax.plot(pde_mean["s"], pde_mean[name], "r--", lw=1.5, label="PDE mean shock frame")
        if pde_local is not None:
            ax.plot(pde_local["s"], pde_local[name], color="tab:orange", ls=":", lw=2.0,
                    label="PDE local shock frame")
        ax.set_ylabel(lab)
        ax.set_xlim(*xlim)
    axes[0].set_title(f"Step 7: PDE final profile vs ZND reference {title_suffix}")
    axes[-1].set_xlabel(r"$s=x_{shock}-x$")
    axes[0].legend(loc="best")
    plt.tight_layout()
    plt.show()


def plot_step7_lambda_focus(pde_mean, pde_local, znd_cj, znd_speed=None, xlim=(0.0, 80.0)):
    plt.figure(figsize=(10, 4.8))
    plt.plot(znd_cj["s"], znd_cj["lam"], "k-", lw=2.0, label="CJ-ZND")
    if znd_speed is not None:
        plt.plot(znd_speed["s"], znd_speed["lam"], "b-.", lw=1.5, label="ZND at D_PDE")
    plt.plot(pde_mean["s"], pde_mean["lam"], "r--", lw=1.5, label="PDE mean shock frame")
    if pde_local is not None:
        plt.plot(pde_local["s"], pde_local["lam"], color="tab:orange", ls=":", lw=2.0,
                 label="PDE local shock frame")
    plt.xlim(*xlim)
    plt.xlabel(r"$s=x_{shock}-x$")
    plt.ylabel(r"$\lambda$")
    plt.title("Step 7: focused reaction-zone comparison")
    plt.legend(loc="best")
    plt.grid(True, alpha=0.25)
    plt.tight_layout()
    plt.show()


def plot_step7_front_positions(x, y, snapshots, par: Params, p_level=2.0, lambda_level=0.5):
    fig, axes = plt.subplots(1, 2, figsize=(13, 5), sharey=True)
    for t, Uinner in snapshots:
        rho, u, v, p, lam = cons_to_prim_2d(Uinner, par.gamma, par.Q)
        xs = front_positions_by_level_lab(x, p, p_level, direction="decreasing", choose="rightmost")
        xl = front_positions_by_level_lab(x, lam, lambda_level, direction="decreasing", choose="rightmost")
        axes[0].plot(xs, y, label=f"t={t:.2f}")
        axes[1].plot(xl, y, label=f"t={t:.2f}")
    axes[0].set_title("p-threshold shock front")
    axes[1].set_title(r"reaction front $\lambda=0.5$")
    axes[0].set_xlabel("x-front")
    axes[1].set_xlabel("x-front")
    axes[0].set_ylabel("y")
    axes[0].legend(loc="best")
    axes[1].legend(loc="best")
    plt.suptitle("Step 7: transverse front positions")
    plt.tight_layout()
    plt.show()


def independent_driver_sweep_2d(par, cj, p_ign_values, xL, xR, yL, yR, Nx, Ny,
                                final_time, cfl, react_cfl, t_fit_min,
                                base_ignition=None):
    print("\n=== Step 7 independent 2D ignition-driver sweep ===")
    print("-" * 128)
    print(f"{'p_ign':>10s} {'D_shock':>14s} {'D_lambda':>14s} {'D_CJ':>14s} {'D_shock-D_CJ':>16s} {'shock_rms':>14s} {'lambda_rms':>14s} {'A_shock_f':>12s}")
    print("-" * 128)
    for p_ign in p_ign_values:
        ign = dict(base_ignition) if base_ignition is not None else {}
        ign["p_ign"] = float(p_ign)

        ign.setdefault("edge_amp", 0.0)
        ign.setdefault("p_amp", 0.0)
        ign.setdefault("v_amp", 0.0)
        ign.setdefault("hotspots", [])
        r = solve_2d_independent_ignition(
            par,
            xL=xL, xR=xR, yL=yL, yR=yR,
            Nx=Nx, Ny=Ny,
            final_time=final_time,
            cfl=cfl,
            react_cfl=react_cfl,
            ignition=ign,
            snapshot_times=[0.0, final_time],
            tracking_dt=0.5,
        )
        Ds, bs, rms_s = fit_mean_front_speed_2d(r["trajectory"], t_min=t_fit_min, front="shock")
        Dl, bl, rms_l = fit_mean_front_speed_2d(r["trajectory"], t_min=t_fit_min, front="lambda")
        dfin = lab_front_diagnostics(r["x"], r["U"], par, p_level=r["p_track_level"], lambda_level=r["lambda_track_level"])
        print(
            f"{p_ign:10.3f} {Ds:14.6f} {Dl:14.6f} {cj['D_cj']:14.6f} "
            f"{Ds - cj['D_cj']:16.6f} {rms_s:14.6e} {rms_l:14.6e} {dfin['shock_front_amp']:12.6f}"
        )
    print("-" * 128)



# Main: independent 2D ignition and ZND comparison


def main():
    par = Params()

    # ---------- independent ZND reference ----------
    cj = find_cj_speed(par)
    prof_cj = cj["profile"]

    print("\n=== ZND reference for Step 7 ===")
    print(f"D_CJ                       = {cj['D_cj']:.12f}")
    print(f"reaction length (10%-90%) = {prof_cj['reaction_length_10_90']:.8f}")
    print(f"reaction length (5%-95%)  = {prof_cj['reaction_length_5_95']:.8f}")
    print(f"s(lambda=0.5)             = {prof_cj['s_lam_05']:.8f}")
    print(f"s(lambda=0.9)             = {prof_cj['s_lam_09']:.8f}")
    print(f"q_end/c_end               = {prof_cj['q_end_over_c_end']:.12f}")

    # ---------- Step 7 setup ----------
    # case_type = "planar"    : 2D analogue of the 1D independent ignition test.
    # case_type = "perturbed" : independent ignition with a corrugated driver edge
    #                           and weak 2D seeds/hot spots.
    case_type = "perturbed"

    xL = -50.0
    xR = 250.0
    yL = 0.0
    yR = 30.0

    dx_target = 0.35
    dy_target = 0.35
    Nx = int(round((xR - xL) / dx_target))
    Ny = int(round((yR - yL) / dy_target))

    final_time = 25.0
    cfl = 0.20
    react_cfl = 0.20
    tracking_dt = 0.25
    t_fit_min = 8.0


    base_ignition = {
        "x_ign_left": -50.0,
        "x_ign_right": 0.0,
        "rho_ign": 1.0,
        "u_ign": 0.0,
        "v_ign": 0.0,
        "p_ign": 55.0,
        "lam_ign": 0.0,
        "smooth": True,
        "smooth_width": 0.7,
    }

    if case_type == "planar":
        ignition = dict(base_ignition)
        ignition.update({
            "edge_amp": 0.0,
            "p_amp": 0.0,
            "v_amp": 0.0,
            "hotspots": [],
        })
    elif case_type == "perturbed":
        ignition = dict(base_ignition)
        ignition.update({
            "edge_amp": 1.50,
            "edge_mode": 1,
            "edge_phase": 0.0,
            "edge_extra_modes": [
                {"mode": 2, "amp": 0.45, "phase": 0.7},
                {"mode": 3, "amp": 0.25, "phase": 1.3},
            ],
            "p_amp": 0.03,
            "v_amp": 0.06,
            "seed_sigma_x": 8.0,
            "seed_center_x": 0.0,
            "seed_mode": 1,
            "seed_phase": 0.0,
            "seed_phase_v": 0.5 * math.pi,
            "hotspots": [
                {
                    "center_x": 5.0,
                    "center_y": yL + 0.33 * (yR - yL),
                    "sigma_x": 5.0,
                    "sigma_y": 3.5,
                    "p_amp": 0.12,
                    "v_amp": 0.04,
                },
                {
                    "center_x": 18.0,
                    "center_y": yL + 0.72 * (yR - yL),
                    "sigma_x": 7.0,
                    "sigma_y": 4.5,
                    "p_amp": 0.08,
                    "v_amp": -0.035,
                },
            ],
        })
    else:
        raise ValueError(f"Unknown case_type={case_type}")

    snapshot_times = [0.0, 2.5, 5.0, 10.0, 15.0, 20.0, final_time]

    result = solve_2d_independent_ignition(
        par,
        xL=xL, xR=xR, yL=yL, yR=yR,
        Nx=Nx, Ny=Ny,
        final_time=final_time,
        cfl=cfl,
        react_cfl=react_cfl,
        ignition=ignition,
        snapshot_times=snapshot_times,
        tracking_dt=tracking_dt,
        p_track_level=2.0,
        lambda_track_level=0.5,
    )

    x = result["x"]
    y = result["y"]
    U = result["U"]
    U_initial = result["U_initial"]
    p_max = result["p_max"]
    T_max = result["T_max"]
    schlieren = result["schlieren"]


    D_shock, b_shock, shock_rms = fit_mean_front_speed_2d(result["trajectory"], t_min=t_fit_min, front="shock")
    D_lambda, b_lambda, lambda_rms = fit_mean_front_speed_2d(result["trajectory"], t_min=t_fit_min, front="lambda")

    final_diag = lab_front_diagnostics(x, U, par, p_level=result["p_track_level"], lambda_level=result["lambda_track_level"])
    initial_diag = lab_front_diagnostics(x, U_initial, par, p_level=result["p_track_level"], lambda_level=result["lambda_track_level"])

    x_shock_final = final_diag["shock_front_mean"]
    rho, u, v, p, lam = cons_to_prim_2d(U, par.gamma, par.Q)
    x_shock_y = front_positions_by_level_lab(x, p, result["p_track_level"], direction="decreasing", choose="rightmost")

    s_max = 120.0
    n_s = 2000
    pde_mean_prof = extract_mean_shock_attached_profile(x, U, par, x_shock_final, s_max=s_max, n_points=n_s)
    pde_local_prof = extract_local_shock_attached_profile(x, y, U, par, x_shock_y, s_max=s_max, n_points=n_s)
    znd_cj_on_s = znd_profile_on_s_grid(prof_cj, pde_mean_prof["s"], par)


    znd_speed_on_s = None
    prof_speed = None
    if np.isfinite(D_shock) and D_shock > cj["D_cj"] + 1.0e-5:
        try:
            prof_speed = build_znd_profile(par, D_shock)
            znd_speed_on_s = znd_profile_on_s_grid(prof_speed, pde_mean_prof["s"], par)
        except Exception as exc:
            print(f"\nWarning: overdriven ZND at D_shock={D_shock:.8f} failed: {exc}")
            prof_speed = None
            znd_speed_on_s = None


    metrics_cj = reaction_zone_metrics_from_profile(znd_cj_on_s)
    metrics_mean = reaction_zone_metrics_from_profile(pde_mean_prof)
    metrics_local = reaction_zone_metrics_from_profile(pde_local_prof)
    metrics_speed = reaction_zone_metrics_from_profile(znd_speed_on_s) if znd_speed_on_s is not None else None

    err_mean_cj = profile_error_report_on_s(pde_mean_prof, znd_cj_on_s, smooth_window=(1.0, 80.0))
    err_local_cj = profile_error_report_on_s(pde_local_prof, znd_cj_on_s, smooth_window=(1.0, 80.0))
    err_mean_speed = profile_error_report_on_s(pde_mean_prof, znd_speed_on_s, smooth_window=(1.0, 80.0)) if znd_speed_on_s is not None else None
    err_local_speed = profile_error_report_on_s(pde_local_prof, znd_speed_on_s, smooth_window=(1.0, 80.0)) if znd_speed_on_s is not None else None

    cellular_diag = {
        "schlieren_max_final": float(np.max(schlieren)),
        "p_max_global_history": float(np.max(p_max)),
        "T_max_global_history": float(np.max(T_max)),
        "pmax_std_y_max": float(np.max(np.std(p_max, axis=1))),
        "Tmax_std_y_max": float(np.max(np.std(T_max, axis=1))),
    }

    # text output
    print("\n=== Step 7: independent 2D ignition and ZND comparison ===")
    print(f"case_type                  = {case_type}")
    print(f"Domain x                   = [{xL:.3f}, {xR:.3f}]")
    print(f"Domain y                   = [{yL:.3f}, {yR:.3f}]")
    print(f"Nx, Ny                     = {Nx}, {Ny}")
    print(f"dx, dy                     = {result['dx']:.8e}, {result['dy']:.8e}")
    print(f"final_time                 = {final_time:.8f}")
    print(f"CFL                        = {cfl:.8f}")
    print(f"react_cfl                  = {react_cfl:.8f}")
    print(f"time steps                 = {result['steps']}")
    print(f"reaction substeps          = {result['reaction_substeps']}")
    print(f"p-track level              = {result['p_track_level']:.8f}")
    print(f"lambda-track level         = {result['lambda_track_level']:.8f}")

    print("\nIndependent ignition parameters:")
    for k, vv in ignition.items():
        print(f"{k:>28s} = {vv}")

    print("\nFront tracking from mean 2D fronts")
    print(f"D_shock from p-threshold   = {D_shock:.8f}")
    print(f"D_lambda from lambda=0.5   = {D_lambda:.8f}")
    print(f"D_CJ from ZND              = {cj['D_cj']:.8f}")
    print(f"D_shock - D_CJ             = {D_shock - cj['D_cj']:.8f}")
    print(f"shock trajectory RMS       = {shock_rms:.8e}")
    print(f"lambda trajectory RMS      = {lambda_rms:.8e}")
    print(f"x_shock_mean(final)        = {final_diag['shock_front_mean']:.8f}")
    print(f"x(lambda=0.5, final)       = {final_diag['lambda05_front_mean']:.8f}")

    print_dict_table("\nInitial 2D transverse/front diagnostics:", initial_diag)
    print_dict_table("\nFinal 2D transverse/front diagnostics:", final_diag)
    print_dict_table("\nStep 7 cellular diagnostics:", cellular_diag)

    print("\nReaction-zone widths / locations")
    print("-" * 96)
    print(f"{'profile':>20s} {'s(lam=.5)':>14s} {'s(lam=.9)':>14s} {'L10-90':>14s} {'L5-95':>14s}")
    print("-" * 96)
    def _row(label, m):
        print(f"{label:>20s} {m['s_lam_05']:14.6f} {m['s_lam_09']:14.6f} {m['L10_90']:14.6f} {m['L5_95']:14.6f}")
    _row("CJ-ZND", metrics_cj)
    _row("PDE mean-frame", metrics_mean)
    _row("PDE local-frame", metrics_local)
    if metrics_speed is not None:
        _row("ZND(D_shock)", metrics_speed)
    print("-" * 96)

    print_error_block("\nErrors: PDE mean-shock-frame profile vs CJ-ZND:", err_mean_cj)
    print_error_block("\nErrors: PDE local-shock-frame profile vs CJ-ZND:", err_local_cj)
    if znd_speed_on_s is not None:
        print(f"\nOverdriven-ZND comparison at D_shock = {D_shock:.8f}")
        print_error_block("\nErrors: PDE mean-shock-frame profile vs ZND(D_shock):", err_mean_speed)
        print_error_block("\nErrors: PDE local-shock-frame profile vs ZND(D_shock):", err_local_speed)

    print("\nGrowth/decay indicators:")
    print("-" * 96)
    for key in ["max_std_p_y", "max_std_lam_y", "max_abs_v", "rms_v", "shock_front_amp", "lambda05_front_amp"]:
        a0 = initial_diag.get(key, np.nan)
        a1 = final_diag.get(key, np.nan)
        ratio = a1 / a0 if np.isfinite(a0) and abs(a0) > 1.0e-14 else np.nan
        print(f"{key:>24s}: initial={a0:14.6e}  final={a1:14.6e}  ratio={ratio:14.6e}")
    print("-" * 96)

    print_step7_history_table(result["history"], stride=4)

    # plots
    plot_step7_lab_snapshots(x, result["snapshots"], par,
                             xlim=(xL, min(xR, final_diag["shock_front_mean"] + 30.0) if np.isfinite(final_diag["shock_front_mean"]) else xR))
    plot_step7_front_trajectory(result["trajectory"], D_cj=cj["D_cj"], t_fit_min=t_fit_min)
    plot_step7_shock_attached_profiles(
        pde_mean_prof, pde_local_prof, znd_cj_on_s, znd_speed=znd_speed_on_s,
        title_suffix=f"at t = {final_time}", xlim=(0.0, 100.0)
    )
    plot_step7_lambda_focus(
        pde_mean_prof, pde_local_prof, znd_cj_on_s, znd_speed=znd_speed_on_s,
        xlim=(0.0, 80.0)
    )
    plot_step7_front_positions(x, y, result["snapshots"], par,
                               p_level=result["p_track_level"], lambda_level=result["lambda_track_level"])
    plot_step5_2d_fields(x, y, U, par,
                         xlim=(max(xL, final_diag["shock_front_mean"] - 110.0),
                               min(xR, final_diag["shock_front_mean"] + 20.0)) if np.isfinite(final_diag["shock_front_mean"]) else (xL, xR))
    plot_step6_schlieren_and_sootfoil(x, y, schlieren, p_max,
                                      xlim=(max(xL, final_diag["shock_front_mean"] - 110.0),
                                            min(xR, final_diag["shock_front_mean"] + 20.0)) if np.isfinite(final_diag["shock_front_mean"]) else (xL, xR))
    plot_step6_cellular_diagnostics(x, y, U, p_max, T_max, schlieren, par,
                                    xlim=(max(xL, final_diag["shock_front_mean"] - 110.0),
                                          min(xR, final_diag["shock_front_mean"] + 20.0)) if np.isfinite(final_diag["shock_front_mean"]) else (xL, xR))


    do_driver_sweep = False
    if do_driver_sweep:
        sweep_ign = dict(base_ignition)
        independent_driver_sweep_2d(
            par=par,
            cj=cj,
            p_ign_values=[50.0, 52.5, 55.0, 57.5, 60.0, 62.5],
            xL=xL, xR=xR, yL=yL, yR=yR,
            Nx=500, Ny=16,
            final_time=18.0,
            cfl=cfl,
            react_cfl=react_cfl,
            t_fit_min=7.0,
            base_ignition=sweep_ign,
        )


if __name__ == "__main__":
    main()
```
